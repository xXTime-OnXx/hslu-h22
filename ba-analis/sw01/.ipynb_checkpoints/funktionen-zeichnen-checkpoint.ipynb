```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
x = np.linspace (-2, 2, 500)
```

```python
y = 2 * x - x ** 3
```

```python
pl.plot (x, y)
pl.axis ('equal')
```

```python
pl.plot (x, 3 * 2 ** x, 'red')
pl.plot (x, 3 * 1.5 ** x, 'blue')
pl.plot (x, 3 * 0.5 ** x, 'magenta')
pl.plot (x, 3 * 1.0 ** x, 'gray')
pl.plot (x, x ** 2, 'black')
pl.plot (x, np.exp (x), 'green')
pl.axis ('equal')
pl.ylim ([0, 5])
```
